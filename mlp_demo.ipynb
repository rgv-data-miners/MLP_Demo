```python
import torch
```

```python
K=32
eq2 = torch.nn.Linear(in_features=2,out_features=K)
eq4 = torch.nn.Linear(in_features=K,out_features=2)
eq3 = torch.nn.ReLU()
```

```python
x= torch.rand((100,2))
```

```python
#Eq 2
x = eq2(x)
x = eq3(x)
x = eq4(x)
output = torch.nn.functional.softmax(x)
```

```python
output
```

```python
loss_func = torch.nn.CrossEntropyLoss()

label = torch.cat((torch.zeros(50,1),torch.ones(50,1))).squeeze().long()

```

```python
loss_func(output,label)
```

```python
import torch.nn as nn

class MLP(nn.Module):
  def __init__(self, K):
    super(MLP,self).__init__()
    self.K = K
    self.eq2 = torch.nn.Linear(in_features=2,out_features=K)
    self.eq4 = torch.nn.Linear(in_features=K,out_features=2)
    self.eq3 = torch.nn.ReLU()
  def forward(self,x):
    x = self.eq2(x)
    x = self.eq3(x)
    x = self.eq4(x)
    y = torch.nn.functional.softmax(x)
    return y

model = MLP(64)


```

```python
import torch.optim as optim

optimizer = optim.SGD(model.parameters(), lr=0.01)

for i in range(10000):
  optimizer.zero_grad()
  o = model(x)
  l = loss_func(o,label)
  l.backward()
  optimizer.step()
  print(l)



```
